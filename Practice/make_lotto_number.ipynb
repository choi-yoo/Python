### 로또 번호 생성기

- input 함수로 사용자 입력을 받아, 총 6개의 숫자를 만들어줍니다.

- 사용자 입력은 모든 자연수가 가능하며, 아무 자연수 6개를 골라서 콤마(',')를 기준으로 순서대로 입력합니다.

- 입력은 한번만 받으며, 들어온 입력을 처리해서 모든 숫자가 1 ~ 45 사이의 숫자가 되는 코드를 작성해보세요.

- 코드는 주어진 함수 make_lotto_number를 완성하는 것으로 합니다.

- 자세한 예시는 아래 실행 예시를 참고하세요.



> **만약에, 입력에 대해서 생성한 6개의 번호중에 중복이 존재한다면 둘 중에 큰 숫자에 +1을 해서 다르게 만들어줍니다.
> 이 때, +1을 해서 45를 넘어가게 된다면 숫자 1로 변환해줍니다.**
> 중복되는 값이 2개 이상인 경우에 앞에서부터 순서대로 +1씩 해주면 됩니다. 만약에 6개 모두 중복이 경우에는(확률이 극히 낮습니다만)
> 출력        >>> [22, 22, 22, 22, 22, 22] 에서
> (수정된)출력 >>> [22, 23, 24, 25, 26, 27]로 변경되어서 출력되어야 합니다

```python
def make_lotto_number(string):
    # 문자열을 리스트 numbers로 저장
    numbers = [int(number) for number in string.split(",")]
    
    for i in range(6):
        if numbers[i] > 45:
            numbers[i] %= 46
    
    # 중복된 숫자 처리
    seen = set()
    new_numbers = []
    
    for number in numbers:
        # number가 이미 seen에 있는 경우
        while number in seen:
            number = (number % 45) + 1
        seen.add(number)
        new_numbers.append(number)

    return new_numbers

  
    
# 사용자로부터 6개의 숫자를 쉼표로 구분하여 입력 받아 문자열로 저장
userInput = input("Input 6 numbers separated by commas: ")

lotto_numbers = make_lotto_number(userInput)
print(lotto_numbers)
```
